```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data=pd.read_csv(r"C:\Users\sudhe\Downloads\archive\creditcard.csv")
```

```python
credit_card_data.head()
```

```python
#data analysis
#types of data and contents of imported dataset(dataframe)
credit_card_data.info()
```

```python
#sum of null values
credit_card_data.isnull().sum()
```

```python
#check distribution of legit transaction and fraudulent transaction
#0 represents legit transaction
#1represents fraudulent transaction
credit_card_data['Class'].value_counts()
```

```python
#separating two types of transactions for further analysis
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
legit.shape
```

```python
fraud.shape
```

```python
#stastical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both legit and fraud
credit_card_data.groupby('Class').mean()
```

```python
#under sampling method
#build a sample dataset containing similar distribution of normal transactionand fraud transaction
#number of fraudulent transaction is 492
#number of legit transaction 284315
#picking random 492 points  from legit and joining with fraud
legit_sample=legit.sample(n=492)
```

```python
#joining legit_sample and fraud
#all the 492 fraud will be below legit_sample points
new_dataset=pd.concat([legit_sample,fraud],axis=0)
new_dataset.tail()
```

```python
#this shows that we have both fraud and legit with equal data counts i.e.492
new_dataset['Class'].value_counts()
```

```python
#the columns v1 of below output and
#the output we got by groupby() above tells that
#nature of dataset is almost same
new_dataset.groupby('Class').mean()
```

```python
#splitting data into features and targets
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
y
```

```python
#split data into training data and testing data
x_train,x_test, y_train , y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
print(x.shape,x_train.shape,x_test.shape)
```

```python
#model training
#logistic regression model
model=LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(x_train,y_train)
```

```python
#model evaluation based on accuracy score
#accuracy score on training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy on training data : ',training_data_accuracy)
```

```python
#accuracy on test data
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy score on test data : ',test_data_accuracy)
```
